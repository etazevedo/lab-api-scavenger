```python
url = 'https://api.github.com/repos/ironhack-datalabs/bcn-feb-2019/forks' 
response = requests.get(url, auth = HTTPBasicAuth('etazevedo', key))
```

```python
response.status_code
```

```python
fork_data = pd.DataFrame(pd.json_normalize(response.json()))
fork_data.head()
```
